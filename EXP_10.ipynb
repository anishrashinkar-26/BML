```python
# PCA Step-by-Step in Google Colab

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Sample dataset
X = np.array([
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9]
])

# Step 1: Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Step 2: Covariance matrix
cov_matrix = np.cov(X_std.T)

# Step 3: Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Sort eigenvalues and eigenvectors
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Step 5: Project data
X_pca = X_std.dot(eigenvectors)

# Explained variance
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)

# Results
print("Standardized Data:\n", X_std)
print("\nCovariance Matrix:\n", cov_matrix)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)
print("\nExplained Variance Ratio:\n", explained_variance_ratio)
print("\nTransformed Data (PCA):\n", X_pca)

# Optional: Plot original vs PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_std[:, 0], X_std[:, 1], label='Original Data')
plt.scatter(X_pca[:, 0], np.zeros_like(X_pca[:, 0]), label='PCA Projection', color='red')
plt.legend()
plt.title("PCA Projection")
plt.grid(True)
plt.show()
```
